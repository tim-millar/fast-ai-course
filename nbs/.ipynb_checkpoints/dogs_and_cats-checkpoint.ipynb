```python
#path = "../data/dogscats/sample"
path = "../data/dogscats"
```

```python
from __future__ import division,print_function
import os, json
from glob import glob
import numpy as np
np.set_printoptions(precision=4, linewidth=100)
from matplotlib import pyplot as plt
import utils; reload(utils)
from utils import plots
```

```python
# finetune model and fit model on data
from vgg16 import Vgg16
vgg = Vgg16()
batch_size = 64
batches = vgg.get_batches(path + '/train', batch_size=batch_size)
validation_batches = vgg.get_batches(path + '/valid', batch_size=batch_size)
vgg.finetune(batches)
vgg.fit(batches, validation_batches, nb_epoch=3)
```

```python
# example of data and one hot encoded labels
display_batches = vgg.get_batches(path+'/train', batch_size=4)
imgs,labels = next(display_batches)
plots(imgs, titles=labels)
```

```python
# predictions for example images above
vgg.predict(imgs)
```
